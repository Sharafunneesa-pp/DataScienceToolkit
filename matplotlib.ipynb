### make simple plot

```python
# pyplot is a module in matplotlib package ,which provide simple functions to add plot elements like lines,images,text,etc to the current axes in the current figure
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("title")
plt.show()
```

### format string: [marker][line][color] 

```python
# plt.plot([2,3,4,5],"r",linewidth=10)
plt.plot([2,3,4,5],"r+",markersize=10)
```

```python
plt.plot([1,2,3,4],[10,20,30,40],'o--g')
```

```python
a=np.arange(1,10,2)
plt.plot(a,"ms--",a**2,"yp:",antialiased=True,linewidth=5)
```
