### make simple plot

```python
# pyplot is a module in matplotlib package ,which provide simple functions to add plot elements like lines,images,text,etc to the current axes in the current figure
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("title")
plt.show()
```

### format string: [marker][line][color] 

```python
# plt.plot([2,3,4,5],"r",linewidth=10)
plt.plot([2,3,4,5],"r+",markersize=10)
```

```python
plt.plot([1,2,3,4],[10,20,30,40],'o--g')
```

### keyword argument 

```python
a=np.arange(1,10,2)
plt.plot(a,color="green",antialiased=True,linewidth=5)
```

```python
# dashes=[length,space]
plt.plot([6,7,8,9],dashes=[4,6])
```

```python
# linestyle or ls
plt.plot([6,7,8,9],linestyle="--")
```

```python
# marker:to display break points
plt.plot([6,7,8,9],marker="+",markeredgecolor="red")
```

```python
plt.plot([6,7,8,9],marker="+",markeredgecolor="green",markeredgewidth=20)
```

```python
# markersize
plt.plot([6,7,8,9],marker="+",markeredgecolor="green",markersize=20)
```

```python
# markerfacecolor
plt.plot([6,7,8,9],marker=".",markersize=20,markerfacecolor="red")
```

```python
# markevery:display marker at every mentioned point
plt.plot([6,7,8,9],marker="+",markevery=2)
```

```python
#alpha:transparent,zorder
plt.plot([1,2,3],[4,5,6],alpha=0.1,linewidth=10)  



```

```python
# change the order of the plot
plt.plot([0,0],[-1,1],linewidth=10,zorder=3)  
plt.plot([-1,1],[-1,1],linewidth=10,zorder=2)  
plt.plot([-1,1],[1,-1],linewidth=10,zorder=1) 
plt.show()
```

```python
# xlabel:name to x axis
# xlabel(xlabel,ylabel
# fontdict=None(change the color style of text),labelpad=None(distance b/w x axis and text),**kwargs(change the text properties))
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("x label",{"color":"red"},labelpad=20,size=20)
```

```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.ylabel("y label",{"color":"red"},labelpad=20,size=20)
```

```python
# font properties
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("x label",{"color":"red","size":30,"backgroundcolor":"orange"},labelpad=20)
plt.ylabel("y label",{"backgroundcolor":"red"},labelpad=20,size="small",color="green",alpha=0.3)
```

```python
# fontstyle{normal,italic,oblique}
# fontweight{0-1000 or ultralight,light,normal,regular,book,medium,roman,semibold,demibold,bold,demi,heavy,extra bold,black}
# font-family{serif,sans-serif,cursive,fantasy,monospace}
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("x label",{"color":"red","size":30},labelpad=20,style="italic",)
plt.ylabel("y label",labelpad=20,size="small",color="green",style="oblique",family="fantasy")
```

```python
# title
# title(label,fontdict=None,loc=None(string value allow to place the title at mentioned position;left,right,center)by default it center,pad=None(distance),**kwargs)
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("x label",{"color":"red","size":30},labelpad=20,style="italic",)
plt.ylabel("y label",labelpad=20,size="small",color="green",style="oblique",family="fantasy")
plt.title("title",{"color":"red","size":30},loc="left",pad=20)
```

### grid

```python
# grid((true,falkse,none:by default its none),which='major'(major,minor or both),axis='both'(x,y,both),**kwargs)
plt.plot([1,2,3,4],[10,20,30,40])
plt.xlabel("numbers")
plt.ylabel("values")
plt.title("multiplication of 10")
plt.minorticks_on()
plt.grid(color="red",which='minor',axis="x")
plt.show()
```

```python
# set limit of the axis
# get (xlim()) or set the limits of the current axes(plot)
# xlim(left,right)
# ylim(top,bottom)
plt.plot([1,2,3,4])
left,right=plt.xlim(1,4)
# plt.xlim((1,4))
# plt.xlim(left=5)
plt.ylim(1,4)

```

### legend function 

```python
#legend(*args,**kwargs)
#we can call like legend(),legend(labels),legend(handles,labels)
#automatically detect which element to show
a=np.arange(5,10)
# 1. method
# plt.plot(a**2,a,label="square")
# plt.plot(a**3,a,label="cube")
# plt.xlabel("values")
# plt.ylabel("numbers")
# plt.legend()
# 2. method:- drawback:if only one argument in list, it is first one.
plt.plot(a**2,a)
plt.plot(a**3,a)
plt.xlabel("values")
plt.ylabel("numbers")
plt.legend(["squares","cubes"])
plt.show()


```

```python
# adjust location:loc
# 'best' - Automatically choose the best location
# 'upper right' - 1
# 'upper left' - 2
# 'lower left' - 3
# 'lower right' - 4
# 'right' - 5
# 'center left' - 6
# 'center right' - 7
# 'lower center' - 8
# 'upper center' - 9
# 'center' - 10
a = np.array([1, 2, 3, 4])
plt.plot(a**2, a,label="square")
plt.plot(a**3, a,label="cube")
plt.xlabel("values")
plt.ylabel("numbers")
# plt.legend(["squares", "cubes"], ncol=2,loc=9,columnspacing=8,)  # Legend placed in upper-left corner
# tuple of 2 float
# plt.legend(loc=(1/4,1/4)) 
# bbox_to_anchor(x,y,width,height)
# plt.legend(loc="upper right",bbox_to_anchor=(0.5,0.5,0.5,0.5)) 
# plt.legend(loc=(1,1.1))
# plt.tight_layout()
plt.show()

```

```python
# fontsize
# 'xx-small': 8
# 'x-small': 10
# 'small': 12
# 'medium': 14
# 'large': 16
# 'x-large': 18
# 'xx-large': 20
# fontcolor
# 'black': (0, 0, 0)
# 'red': (1, 0, 0)
# 'green': (0, 1, 0)
# 'blue': (0, 0, 1)
# 'cyan': (0, 1, 1)
# 'magenta': (1, 0, 1)
# 'yellow': (1, 1, 0)
# 'white': (1, 1, 1)
# markerfirst,shadow,frameon,legend title,title fontsize,borderpad,labelspacing
a = np.array([1, 2, 3, 4])
plt.plot(a**2, a,label="square")
plt.plot(a**3, a,label="cube")
plt.xlabel("values")
plt.ylabel("numbers")
leg=plt.legend(title="legend title",title_fontsize=10,borderpad=3,labelspacing=2,markerfirst=False,shadow=True,frameon=True,fontsize="small",edgecolor="green",facecolor="yellow",framealpha=0.5)
# add more legend
plt.gca().add_artist(leg)
leg2=plt.legend(title="legend title")
plt.gca().add_artist(leg2)
plt.tight_layout
```

```python
# xticks() and yticks()
# tick_params(axis="both",**kwargs)
plt.plot([1,2,3,4],[5,6,7,8],)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("plotting points")
plt.grid()
# right,top,bottom,left
# labelsize="xx-small",labelcolorgrid_linewidth,colors="green",labelright="True",labelrotation,grid_color,grid_alpha,grid_linewidth,grid_linestyle
plt.tick_params(right=True,top=True,axis="x",length=20,direction="inout",which="both",width=5,color="red")

# plt.xticks([2,4],["a","b"])
# plt.yticks([3,6,9],["a","b","c"],rotation="vertical",color="green")
plt.show()
```
